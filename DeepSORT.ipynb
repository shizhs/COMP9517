```python
import math
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import os
from deep_sort import tracker, nn_matching, detection
```

```python
metric=nn_matching.NearestNeighborDistanceMetric("euclidean", 5)
max_age=2
n_init=2
tracker=tracker.Tracker(metric, max_age, n_init)
```

```python
def draw_bounding_box(new_img, new_next_cells):
    for i in range(len(new_next_cells)):
        label, cells, y1, y2, x1, x2=new_next_cells[i]
        plt.title("Number of Cells: "+str(len(new_next_cells))+"Max: ")
        plt.imshow(new_img, interpolation="nearest")
        rect = Rectangle((x1, y1), (x2-x1), (y2-y1), fill=False, color='red')
        plt.axes().add_patch(rect) 
        plt.text(x1+1,y1+1, str(label))
    plt.show()  
    return
```

```python
def tracking(image_path, info_path):
    image_names=sorted(os.listdir(image_path))
    for path in image_names:
        if path.endswith(".npy"):
            continue
        img=cv2.imread(image_path + path,0)
        dim = (500, 500)
        img = cv2.resize(img, dim)
        curr_cells=np.load(info_path+path.split('.')[0]+'.npy', allow_pickle=True)
        detections=[]
        for cell in curr_cells:
            label, cropped, y1, y2, x1, x2 = cell[0], cell[1], cell[2], cell[3], cell[4], cell[5]
            detections.append(detection.Detection([x1, y1, x2-x1, y2-y1], 1, [np.sum(cropped)/np.prod(cropped.shape), np.prod(cropped.shape)]))
        
        tracker.predict()
        tracker.update(detections)
        results=tracker._match(detections)
        print(results)
```

```python
tracking('./fluo/Sequence 1/', './fluo/Sequence 1/')
```
